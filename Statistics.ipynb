```python
import pandas as pd
from pandas import DataFrame
import numpy
import matplotlib.pyplot as plt
import datetime
import seaborn
```

```python
#Reading / using the orginal file not the modified by excel
df=pd.read_csv('Team PY Motor_Vehicle data.csv')
df.head(5)
```

```python
#Replace the coloumns head:
df.rename(columns = {'ACCIDENT DATE':'DATE','ACCIDENT TIME':'TIME','CONTRIBUTING FACTOR VEHICLE 1':'CONTRIBUTING FACTOR','VEHICLE TYPE CODE 1':'VEHICLE TYPE'}, inplace = True) 

df.columns
```

```python
# Cleaned DataSet:

df2=pd.DataFrame(df,columns=['DATE','TIME','BOROUGH','CONTRIBUTING FACTOR','VEHICLE TYPE','NUMBER OF PERSONS INJURED','NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED','NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED'])
df2.head()
```

```python
#length of Data and coloumns we are using

print(len(df2))
print(len(df2.columns))
```

```python
df2.dtypes
```

```python
#Q1) Which of the five boroughs has the highest collision in july?


df2['sum-up of PERSONS incidents']=df2['NUMBER OF PERSONS INJURED']+df2['NUMBER OF PERSONS KILLED']
df2['sum-up of PEDESTRIAN incidents']=df2['NUMBER OF PEDESTRIANS INJURED']+df2['NUMBER OF PEDESTRIANS KILLED']
df2['sum-up of CYCLIST incidents'] = df2['NUMBER OF CYCLIST INJURED'] + df2['NUMBER OF CYCLIST KILLED']
df2['sum-up of MOTORIST incidents']= df2['NUMBER OF MOTORIST INJURED']+df2['NUMBER OF MOTORIST KILLED']

#Summed All incidents into one whole incident
df2['Total Incidents'] = df2['sum-up of PERSONS incidents']+df2['sum-up of PEDESTRIAN incidents']+df2['sum-up of CYCLIST incidents']+df2['sum-up of MOTORIST incidents']

#Put All the Rows Together
df2[['BOROUGH','sum-up of PERSONS incidents','sum-up of PEDESTRIAN incidents','sum-up of CYCLIST incidents','sum-up of MOTORIST incidents','Total Incidents']].head(10)

df2.groupby(['BOROUGH'])['Total Incidents'].sum()
```

```python
#Q2) Which Borough has the highest pedestrian injury for the month of July?

df2.groupby('BOROUGH')['NUMBER OF PEDESTRIANS INJURED'].sum()



```

```python
#Q3) Which borough has the highest count for pedestrian injury or death due to intoxicated drivers?


data2=df2.loc[df2['CONTRIBUTING FACTOR'].str.contains('Alcohol Involvement',na = False)]
data2.groupby('BOROUGH')['CONTRIBUTING FACTOR'].count()
```

```python
#Q4) Which borough has the most collisions due to Passing or Improper Lane Usage?
dc={}
for index, row in df2.iterrows():
    if row['CONTRIBUTING FACTOR'] == 'Passing or Lane Usage Improper':
        dc[index]=[row['CONTRIBUTING FACTOR'], row['BOROUGH']]
df3=pd.DataFrame.from_dict(dc, orient='index')
df3.columns=['CONTRIBUTING FACTOR','BOROUGH']
df3.groupby('BOROUGH')['CONTRIBUTING FACTOR'].value_counts()

```

```python
#Q5) Which vehicle Type has the highest collision rate?

df2['sum-up of PERSONS incidents']=df2['NUMBER OF PERSONS INJURED']+df2['NUMBER OF PERSONS KILLED']
df2['sum-up of PEDESTRIAN incidents']=df2['NUMBER OF PEDESTRIANS INJURED']+df2['NUMBER OF PEDESTRIANS KILLED']
df2['sum-up of CYCLIST incidents'] = df2['NUMBER OF CYCLIST INJURED'] + df2['NUMBER OF CYCLIST KILLED']
df2['sum-up of MOTORIST incidents']= df2['NUMBER OF MOTORIST INJURED']+df2['NUMBER OF MOTORIST KILLED']

df2.groupby(['VEHICLE TYPE'])['sum-up of PERSONS incidents','sum-up of PEDESTRIAN incidents','sum-up of CYCLIST incidents','sum-up of MOTORIST incidents'].sum()
```
